```python
import numpy as np
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
actual=np.array(['Cat','Cat','Cat','Not Cat','Cat','Not Cat','Cat','Cat','Not Cat','Cat','Not Cat','Not Cat'])
predicted=np.array(['Cat','Not Cat','Cat','Not Cat','Cat','Cat','Cat','Not Cat','Cat','Cat','Not Cat','Not Cat'])
```

```python
cm=confusion_matrix(actual,predicted)
sns.heatmap(cm, annot=True,fmt=".1f",xticklabels=['Cat','Not Cat'],yticklabels=['Cat','Not Cat'],
            linewidths=3,linecolor="red",cmap='crest',cbar=True)
plt.xlabel('Prediction',fontsize=13)
plt.ylabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
from sklearn import metrics
cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=['Cat','Not Cat'])
cm_display.plot(cmap='crest')
plt.xlabel('Prediction',fontsize=13)
plt.ylabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
print(classification_report(actual,predicted))
```
