```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
d = pd.read_csv('User_Data.csv')
print(d)
```

```python
d.drop(['Gender','User ID'],axis=1,inplace=True)
print(d)
```

```python
x=d.iloc[:,[0,1]].values
y=d.iloc[:,2].values
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,shuffle=True)
print(x_train.shape)
print(x_test.shape)
knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=1)
print(knn.fit(x_train,y_train))
print(knn.score(x_test,y_test))
```

```python
y_predict=knn.predict(np.array([[36,40000]]))
print(y_predict)
```

```python
d = pd.read_csv('User_Data.csv')
d.drop(['User ID'],axis=1,inplace=True)
d.replace({"Male":1,"Female":0},inplace=True)
x=d.iloc[:,[0,1,2]].values
y=d.iloc[:,3].values
print(d)
print(x)
print(y)
```

```python
x=scale(x)
plt.scatter(x[:,1],x[:,2],c=y)
print(x)
```
